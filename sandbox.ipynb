```python
from core.game import Checkers
from core.state_encoder import StateEncoder
from core.model import ResNet
from core.mcts import MCTS
from training.trainer import Trainer
import torch, torch.optim as optim
```

```python
game = Checkers()
encoder = StateEncoder(game)
model = ResNet(input_channels=encoder.get_num_channels(), action_size=game.action_size,
               board_size=(game.row_count, game.col_count), num_resBlocks=3, num_hidden=64,
               device=torch.device('cpu'))
args = { 'num_iterations': 100, 'num_self_play_iterations': 10, 'num_epochs': 2, 'batch_size': 32, 'num_searches': 50 }
mcts = MCTS(game, args, model, encoder)
trainer = Trainer(model, optim.Adam(model.parameters(), lr=1e-3), game, mcts, encoder, args)
# trainer.train(model_dir='models', resume_from=None)
```
